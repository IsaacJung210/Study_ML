```python
with open('student_info.csv','r') as f:
    for line in f:
        (name, height, weight) = line.strip().split(",")
        
        if (not name) or (not height) or (not weight):
            continue
        bmi = int(weight) / (int(height) * int(height))
        result = ''
        if 25 <= bmi:
            result = '과체중'
        elif 18.5 <= bmi:
            result = '정상 체중'
        else: 
            result = '저체중'
        print('\n'.join([
            "이름 : {}",
            "키 : {}",
            "몸무게 : {}",
            "BMI : {}",
            "결과 : {}"
        ]).format(name,height,weight,bmi,result))
        print()
```

```python
## 1~10000 까지의 숫자 중에서
## random 하게 3자리의 수를 100개만 추출해서 data 라는 리스트를 만들고
## input 함수를 사용해서 수치를 입력받아 타입이 다르면 타입에러 ==> "타입에러"
## 선택한 숫자가 data에 없으면 인덱스에러 ==> "인덱스에러"
## 선택한 숫자가 있는 곳의 index 번호를 반환
import random
data_tmp = list (range(101,1000))
data = []
for i in range(100):
    tmp = random.choice(data_tmp)
    data.append(tmp)
```

```python
input_tmp = input("데이터를 입력하세요")
try:
    input_tmp = int(input_tmp)
    ind = data.index(input_tmp)
    print(ind)
except TypeError as e:
    print("틀린 타입 : ", type(e))
    print("틀린 내용 : ", e)
except ValueError as e:
    a = random.choice(range(100))
    print(f'{data[a]},data 리스트의 {a}의 위치에 있습니다.')
```
