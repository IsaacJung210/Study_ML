```python
continent = pd.read_sql('select distinct continent from world',conn)
continent_options = [{"label":list(i)[0],"value":list(i)[0]} for i in continent.values]
```

```python
app = Dash(__name__)
app.layout = html.Div(
    [
        html.Label("대륙"),
        dcc.Dropdown(id="dropdown1",
                     options = continent_options),
        html.Label('국가'),
        html.Div(id="output-dropdown"),
        dcc.RadioItems(id="radio-items",
                       options = ['파이그래프','막대그래프','라인그래프'], 
                       value = "막대그래프",
                       inline = True),
        html.Hr(),
        html.Div(id="output-country")
    ]
)
```

```python
@app.callback(
    Output('output-dropdown','children'),
    Input("dropdown1",'value')
    
)
def updata_country_dropdown(continent):
    country = pd.read_sql(f"select distinct country from world where continent = '{continent}'",conn)
    country_options = [{"label":list(i)[0],"value":list(i)[0]} for i in country.values]
    return html.Div(
        [
            dcc.Dropdown(
                id="dropdown2",
                options = country_options
            )         
        ]    
    )
@app.callback(
    Output("output-country","children"),
    Input('dropdown1','value'),
    Input('dropdown2','value'),
)
def update_country(x1,x2):
    data = pd.read_sql(f"select year,pop from world where country = '{x2}'",conn)
    pop_data = pd.read_sql(f"select country, pop from world where pop = (select max(pop) from world where continent = '{x1}')",conn)
    
    continent_population = pd.read_sql(f"select pop, country from world where continent = '{x1}' and year = 2007",conn)
    continent_country = continent_population.loc[:,"country"]
    continent_pop = continent_population.loc[:,"pop"]
    colors = ["Gray"]*len(continent_population)
    colors[continent_pop[continent_pop == continent_pop.max()].index[0]] = "Crimson"
    colors[continent_pop.sort_values(ascending=False).index[1]] = "Blue"
    country_name = pop_data.loc[:,"country"][0]
    pop = pop_data.loc[:,"pop"][0]
    
    fig = go.Figure([go.Bar(x=continent_country.values,y=continent_pop.values,marker_color=colors)])
    fig1 = px.bar(data,x='year',y='pop')
    fig2 = px.line(data,x='year',y='pop')
    fig3 = px.pie(data,values="pop",names="year")
    return html.Div([
        html.Div(f'{x1}대륙에서 인구가 가장 많은 나라는 {country_name}이고 인구수 {pop}니다.'),
        dcc.Graph(figure = fig),
        html.Div(f'{x1}대륙의 {x2}나라의 년도별 인구수 그래프입니다.'),
        html.Hr(),
        dcc.Graph(id="bar-graph",
                  figure = fig1),
        dcc.Graph(id="line-graph",
                  figure = fig2),
        dcc.Graph(id="pie-graph",
                  figure = fig3)
    ])
```

```python
app.run_server(host=host,port="7777")
```
