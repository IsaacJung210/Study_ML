```python
def 복원(x,n):
    result = []
    while True:
        a = random.choice(x)
        if a not in result:
            result.append(a)
            if len(result) == n:
                break    
        else:
            pass
    return result
```

```python
def welcome_n(name,n):
    x = name.copy()
    result = []
    while True:
        choice = random.choice(x)
        result.append(choice)
        if len(result) == n:
            break
        x.remove(choice)
    return result
```

```python
def welcome_n1(name,n):
    x = name.copy()
    result = []
    cnt = 0
    while cnt < n:
        choice = random.choice(x)
        result.append(choice)
        x.remove(choice)
        cnt += 1
    return result
```

```python
welcome_n1(name,5)
```

```python
def welcome_n2(name,n):
    x = name.copy()
    result = []
    for i in range(n):
        choice = random.choice(x)
        result.append(choice)
        x.remove(choice)
    return result
```

```python
welcome_n2(name,4)
```

```python
def welcome_n3(name,n):
    x = name.copy()
    result = []
    if len(x) < n:
        print(f'총 학생의 수는 {len(x)}명인데 입력한 학생 수는 {n}명입니다.')
    for i in range(n):
        choice = x[random.choice(range(len(x)))]
        result.append(choice)
        x.remove(choice)
    return result
```

```python
welcome_n3(name,5)
```

```python
def fibo2(n):
    f_1 = 1
    f_2 = 1
    tmp = [f_1,f_2]
    for i in range(n):
        tmp.append(tmp[i]+tmp[i+1])
    return tmp[-3]
```

```python
def flatten(x):
    result = []
    for i in x:
        if type(i) == int:
            result += [i]
        else:
            result += flatten(i)
    return result
```
